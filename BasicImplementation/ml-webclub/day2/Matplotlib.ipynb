## Matplotlib

''Matplotlib'' is probably the single most used Python package for 2D-graphics. It provides both a very quick way to visualize data. So when you are working with a dataset you would want to see data visually to get a better idea of the data or display the data to convey their results to someone.


```python
import matplotlib
import pandas as pd
```

#### Pyplot

matplotlib.pyplot is a collection of functions that make matplotlib work. Each pyplot function makes some change to a figure: e.g., creates a figure, creates a plotting area in a figure, plots some lines in a plotting area, decorates the plot with labels, etc.
In this example we are plotting points - (1,1),(2,4),(3,9),(4,16)
For every x, y pair of arguments, there is an optional third argument which is the format string that indicates the color and line type of the plot. So here ro is red 'o' which represents the point

```python
import matplotlib
import matplotlib.pyplot as plt
plt.plot([1,2,3,4], [1,4,9,16], 'ro')
plt.axis([0, 6, 0, 20])
plt.show()
```

```python
import numpy as np

# evenly sampled at 0.1 interval
t = np.arange(0,2,0.1)

# red dashes, blue squares and green triangles
plt.plot(t, t, 'r--', t, t**2, 'bs', t, t**3, 'g^')
plt.show()
```

So in this example we have three equations y=x, y=x^2 , y=x^3 and they are plotted above

#### Descriptive charts

We use barplots , scatterplots , histogram generally for visualising data because its more descriptive and conveys more info. So Ill introduce you guys to barplot

```python
x = [5,8,10] 
y = [12,16,6]  
plt.bar(x, y, align = 'center') 
plt.title('Bar graph') 
plt.ylabel('Y axis') 
plt.xlabel('X axis')  

plt.show()
```

In real world we would be plotting huge datasets and not just an array of numbers. So with the help of pandas we can load the dataset and plot it using matplotlib

```python
housing=pd.read_csv('housing.csv')
housing.plot(kind="scatter",x='longitude',y='latitude')
plt.show()
```
