```python
# handling missing data
(X['Economy..GDP.per.Capita.'] == 0).sum() # No.of zeros in this column  
```

```python
X[['Economy..GDP.per.Capita.']] = X[['Economy..GDP.per.Capita.']].replace(0, np.NaN) # replacing 0's in the column with NaN
```

```python
X[['Economy..GDP.per.Capita.', 'Generosity', 'Health..Life.Expectancy.']] = X[['Economy..GDP.per.Capita.', 'Generosity', 'Health..Life.Expectancy.']].replace(0, np.NaN)
# performing the same action on multiple columns
```

```python
(X['Economy..GDP.per.Capita.'] == 0).sum()
```

```python

X = X.replace(0, np.NaN) # if wkt 0 implies invlid observation throughout the dataset
# the same action is performed on the entire dataset
```

```python
# Some algorithms donot work with NaN values, like LDA. So we may have to remove such records or impute them(Change their values)
```

```python
copy1 = X
copy2 = X # making two copies for two different approaches
```

```python
copy1.dropna(inplace=True) # deletes all the records with missing data
# this approach may not be feasible if significant number of records have missing data in the table
```

```python
copy1.shape 
```

```python
# imputing of values in training data implies, the test data also needs to be imputed 
# with the same type of value. The value to be imputed can be mean, median or mode, value from a randomly chosen record, etc
```

```python
copy2.fillna(copy2.mean(), inplace=True) # impute methodology
```

```python
copy1.isnull().sum() # verify if no NULL values are present
```
