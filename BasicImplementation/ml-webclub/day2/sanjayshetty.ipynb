```python
import pandas as pd
import numpy as np
import matplotlib 
import matplotlib.pyplot as plt
def costfunction(X,Y,theta):
    m=X.size
    cost=(np.sum((X.dot(theta)-Y)**2)/(2*m))
    return cost


def gradientdescent(X, Y, theta, alpha, iterations):

    cost_history=[0]*iterations
    m=X.size
    for i in range(iterations):
        hypothesis = X.dot(theta)
        loss = hypothesis-Y
        gradient = X.T.dot(loss)/m   #1 by 1 matrix
        theta = theta - alpha*gradient
        cost = costfunction(X, Y, theta)
        cost_history[i] = cost
    print("minimum cost",cost_history[iterations-1])

    return theta, cost_history

def main():
    x=pd.read_csv('train.csv',sep=',')
    
    X=np.ones((2,10000))
    X1=np.array(x['x1'])
    Y=np.array(x['y'])
    X1=X1[0:10000]
    Y=Y[0:10000]
    X1.resize(1,10000)
    Y.resize(10000,1)
    X[1]=X1
    X.resize(10000,2)
    theta=np.zeros((2,1))
    #  X = 10000 * 2   Y=10000*1   THETA=2*1
    
    plt.plot(X.T[1],Y,'ro')
    plt.axis([0,0.2,0,0.2])
    plt.xlabel("x1")
    plt.ylabel('y')
    plt.title("linear regression")
    plt.show()
    it=1000
    theta,cs=gradientdescent(X,Y,theta,0.015,it)
    
    counter=np.arange(1,it+1,1).reshape(1,it)
    cost_his=np.array(cs)
    cost_his.resize(1,it)
    
   
    plt.plot(counter,cost_his,'ro')
    plt.axis([1,it,0,.2])
    plt.xlabel("counter")
    plt.ylabel('cost function')
    plt.title("checking the convergence")
    plt.show()
    x1=input("enter the value of x for which y is to predicted")
    a=float(theta[0])
    b=float(theta[1])
    print("predicted value is",a+b*float(x1))
    
    
main()
```
